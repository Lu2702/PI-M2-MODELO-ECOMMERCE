## PI 1: Análisis de negocio y descubrimiento de requisitos

* Analizar las preguntas de negocio a la luz de los datos presentados en el punto anterior.
* Entender qué quiere resolver el negocio y qué datos están disponibles para ello.

Los stakeholders han definido las siguientes preguntas clave que reflejan las prioridades estratégicas del negocio:

1. ***¿Cuál es la tasa de conversión de usuarios que agregan productos al carrito y concretan una compra?***
2. ***¿Cuál es el ingreso generado por mes y el ingreso total acumulado?***
3. ***¿Qué categoría de producto presenta mayores ingresos?***
4. ***¿Qué usuario genera los mayores ingresos?***
5. ***¿Cual es la tasa de cancelación de ordenes de compra?***

Estas preguntas permiten realizar un análisis con una visión integral del negocio, brindando a esta empresa emergente del sector de comercio eléctronico la base sólida para definir KPIs relevantes, generar insights accionables y establecer metas orientadas al crecimiento sostenido. 
Con base en estas preguntas, se busca implementar un modelo de datos relacional y dimensional que consolide de manera estructurada la información proveniente de distintas áreas del negocio.Esta es la información actualmente disponible:

 - Usuarios: ID, Nombre, Apellido, dni, Email, fecha de registro
 - Productos: ID, Nombre, Descripción, Categoria, Precio unitario, Stock
 - Categorias: ID, Nombre, Descripción
 - Carrito: ID, Usuario, Producto, Cantidad, Fecha agregado
 - Ordenes: ID, Usuario, Fecha orden, Total, Estado
 - Detalle de ordenes: Id, Orden, Producto, Cantidad, Precio unitario
 - Metodos de pago: ID, Nombre, Descripción
 - Ordenes Metodos de pago: Id, Orden, Metodo de pago, Monto pagado
 - Historial de pago: ID, Orden, Metodo de pago, Monto, Fecha de pago, Estado de pago
 - Reseñas productos: ID, Usuario, Producto, Calificación, Comentario, Fecha
 - Direcciones de envio: ID, Usuario, Calle, Ciudad, Departamento, Provincia, Distrito, Estado, Codigo postal, País. 


Este modelo facilitará tanto la exploración analítica como la toma de decisiones estratégicas en los distintos niveles de la organización, permitiendo resolver las preguntas de negocio establecidas.

### Modelar Slowly Changing Dimensions (SCDs) con una estrategia adecuada (SCD Type 1, 2 o 3).

| Tipo SCD | Descripción | Método | ¿Se conserva el historial? | Uso típico |
|----------|-------------|--------|----------------------------|------------|
| Tipo 1 | Se sobrescribe el valor anterior con el nuevo.| UPDATE directo | NO | Correcciones simples |
| Tipo 2 | Se crea una nueva fila para cada cambio. Se conserva la anterior. | Nuevas filas | SI | Trazabilidad completa |
| Tipo 3 | Se guarda el valor anterior en otra columna. Se actualiza también el actual. | Nuevas columnas | Parcial | Historal límitado |
